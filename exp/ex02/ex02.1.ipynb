```python
import os
import sys
import csv
import logging
import matplotlib.pyplot as plt
from sklearn.model_selection import ShuffleSplit
import seaborn as sns

repo_root = os.path.abspath(os.path.join(os.getcwd(), '..', '..',))
sys.path.append(repo_root)

from src import *
```

```python
columns = ['game', 'std', 'stdl', 'ho', 'hol', 'spring_rank', 'spring_rank_l', 'page_rank', 'page_rank_l', 'point']
mean_keys, prop_keys = [f'mean_{columns[i]}' for i in range(1, 10)]

prop_results_all = {key: [] for key in prop_keys}
mean_results_all = {key: [] for key in mean_keys}
```

```python


def plot_distributions(prop_results_all, mean_results_all):
    fig, axes = plt.subplots(1, 2, figsize=(20, 8))
    
    # Plot distributions of proportions using kdeplot
    for key, values in prop_results_all.items():
        sns.kdeplot(values, ax=axes[0], label=key)
    axes[0].set_title('Proportions Distribution')
    axes[0].set_xlabel('Proportion')
    axes[0].set_ylabel('Density')
    axes[0].legend()

    # Plot distributions of means using kdeplot
    for key, values in mean_results_all.items():
        sns.kdeplot(values, ax=axes[1], label=key)
    axes[1].set_title('Means Distribution')
    axes[1].set_xlabel('Mean')
    axes[1].set_ylabel('Density')
    axes[1].legend()
    
    plt.tight_layout()
    plt.show()
```

# Standard 

```python
plot_distributions(prop_results_all, mean_results_all)
```

# Higher Order 

```python
N, M, K1, K2 = 1000, 2000, 5, 5
compared_axis = 3
prop_results_all, mean_results_all = evaluate_models_fixed_train_size(N, M, K1, K2, compared_axis)
```

```python
plot_distributions(prop_results_all, mean_results_all)
```

```python
N, M, K1, K2 = 1000, 2000, 5, 5
compared_axis = 4
prop_results_all, mean_results_all = evaluate_models_fixed_train_size(N, M, K1, K2, compared_axis, leadership=True)
```
