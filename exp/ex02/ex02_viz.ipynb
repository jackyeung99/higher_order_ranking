```python
df = pd.read_csv('ex02_results.csv')
```

```python
grouped = df.groupby(['N', 'M', 'K2', 'model']).mean().reset_index()

ho_data = grouped[(grouped['model'] == 'ho') & (grouped['M'] == 7419)]
hol_data = grouped[(grouped['model'] == 'hol') & (grouped['M'] == 7419)]

# Plot for 'ho' model
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(18, 6), sharey=True)
ax1.plot(ho_data['K1'], ho_data['prop_ho_greater_std'], marker='o', linestyle='-', label='proportion of games ho likelihood > std likelihood')
ax1.plot(ho_data['K1'], ho_data['prop_hol_greater_std'], marker='x', linestyle='--', label='proportion of games hol likelihood > std likelihood')
ax1.set_xlabel('K')
ax1.set_ylabel('Proportion Better')
ax1.set_title('HO synthetic generated model')
ax1.legend()
ax1.grid(True)

ax2.plot(hol_data['K1'], hol_data['prop_ho_greater_std'], marker='o', linestyle='-', label='proportion of games ho likelihood > std likelihood')
ax2.plot(hol_data['K1'], hol_data['prop_hol_greater_std'], marker='x', linestyle='--', label='proportion of games hol likelihood > std likelihood')
ax2.set_xlabel('K')
ax2.set_ylabel('Proportion Better than std model')
ax2.set_title('HOL synthetic generated model')
ax2.legend()
ax2.grid(True)

```

```python

unique_constants = sorted(df['K1'].unique())
grouped = df.groupby(['N', 'M', 'K2', 'model']).mean().reset_index()

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(18, 6), sharey=True)

for hyperedge_size in unique_constants:

    ho_data = grouped[(grouped['model'] == 'ho') & (grouped['K1'] == hyperedge_size)]

    ax1.plot(grouped['M'].unique(), ho_data['prop_ho_greater_std'], label=f'hyperedge size = {hyperedge_size}', alpha=0.7)
    ax2.plot(grouped['M'].unique(), ho_data['prop_hol_greater_std'], label=f'hyperedge size = {hyperedge_size}', alpha=0.7)
   
ax1.set_xlabel('M')
ax1.set_ylabel('Proportion Better than std model')
ax1.set_title('HO Model')
ax1.legend()
ax1.grid(True)

ax2.set_xlabel('M')
ax2.set_title('HOL Model')
ax2.legend()
ax2.grid(True)

fig.suptitle('hyperedge to game comparison for HO synthetically generated data')
```
