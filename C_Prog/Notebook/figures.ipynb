```python
import csv
import matplotlib.pyplot as plt
%matplotlib inline 
import numpy as np
```

```python
data = []
with open('data_ho.txt', newline='') as f:
    rows = csv.reader(f, delimiter=' ', quotechar='|')
    for row in rows:
        tmp = []
        for r in row:
            tmp.append(float(r))
        data.append(tmp)
data_ho = np.array(data)

###
data = []
with open('data_leader.txt', newline='') as f:
    rows = csv.reader(f, delimiter=' ', quotechar='|')
    for row in rows:
        tmp = []
        for r in row:
            tmp.append(float(r))
        data.append(tmp)
data_leader = np.array(data)
```

```python
fig = plt.figure()
fig.set_size_inches(14,5)
plt.rc('text', usetex=False)
plt.rc('font', size=24, **{'family':'DejaVu Sans','sans-serif':['Helvetica']})
plt.rcParams['xtick.major.pad'] = 8
plt.rcParams['ytick.major.pad'] = 8

ax1 = fig.add_subplot(1,2,1)
ax2 = fig.add_subplot(1,2,2)


ax1.plot(data_ho[:,1],data_ho[:,2], marker='o', color = 'red', alpha=0.5, label='ho')
ax1.plot(data_ho[:,1],data_ho[:,3], marker='s', color = 'blue', alpha=0.5, label='std')
ax1.plot(data_ho[:,1],data_ho[:,4], marker='d', color = 'green', alpha=0.5, label='std')


ax1.set_xscale('log')
ax1.set_yscale('log')



ax2.plot(data_leader[:,1],data_leader[:,2], marker='o', color = 'red', alpha=0.5, label='ho')
ax2.plot(data_leader[:,1],data_leader[:,3], marker='s', color = 'blue', alpha=0.5, label='std')
ax2.plot(data_leader[:,1],data_leader[:,4], marker='d', color = 'green', alpha=0.5, label='std')


ax2.set_xscale('log')
ax2.set_yscale('log')
```
